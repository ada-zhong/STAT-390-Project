```python
import pandas as pd
```

```python
bus = pd.read_csv('CTA_BusStops.csv')
bus.head()
```

**Note to self: verify that latitude comes first in the coordinate**

```python
# Extract latitude and longitude from "the_geom" column
bus['latitude'] = bus['the_geom'].str.split(' ').str[1].str.replace('(', '').astype(float)
bus['longitude'] = bus['the_geom'].str.split(' ').str[2].str.replace(')', '').astype(float)

bus.head()
```

```python
# Move columns to the front
bus = bus.reindex(columns=['latitude', 'longitude'] + list(bus.columns[:-2]))
bus.head()
```
