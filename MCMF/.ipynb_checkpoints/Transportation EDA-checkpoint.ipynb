```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mcmf = pd.read_csv('mcmf.csv')
```

```python
mcmf = mcmf.drop('Unnamed: 0', axis=1)
```

```python
mcmf.head()
```

```python
grouped = mcmf.groupby('geographic_cluster_name').agg({
    'km_to_bus_stop': 'mean',
    'km_to_train_stop': 'mean'
}).reset_index()
```

```python
grouped
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped['geographic_cluster_name'], grouped['km_to_bus_stop'], label='Average Distance to Bus Stop')
plt.bar(grouped['geographic_cluster_name'], grouped['km_to_train_stop'], label='Average Distance to Train Stop')
plt.xlabel('Neighborhood')
plt.ylabel('Average Distance (km)')
plt.title('Proximity of Neighborhoods to Bus and Train Stops')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
